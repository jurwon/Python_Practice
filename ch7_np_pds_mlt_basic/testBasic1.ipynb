```python
import numpy as np
import pandas as pd
import matplotlib

list = [[1,2],[3,4],[5,6]]

np_array = np.array(list)

#슬라이싱 , [시작값:끝값:증가값], 끝값 : 미포함
print(f"list[1][1] : {np_array[1][1]}")
print(f"list[0:1] : {np_array[0:1]}") #0행만
print(f"list[0:2] : {np_array[0:2]}") #0~1행
print(f"list[:][1] : {np_array[:,1]}") #전체 행, 1열

print(f"list[:][1] : {np_array[[0,2],[1]]}") #0행,2행의 1열
```

```python
import numpy as np
import pandas as pd
import matplotlib

list = [[1,2],[3,4],[5,6]]

np_array = np.array(list)

#배열 각 요소에 특정 값 한번에 모두 더하기
print(f"np_array \n: {np_array}")
print(f"np_array +100 \n: {np_array+100}")

#각행, 각열의 합, axis=0 열기준, axis=1
print(f"np_array 열의 합 : \n {np.sum(np_array,axis=0)}")

print(f"np_array 행의 합 : \n {np.sum(np_array,axis=1)}")


#각 행, 열의 기준 정렬
print(f"np_array 행 기준 정렬 : \n {np.sort(np_array,axis=1)}")
print(f"np_array 열 기준 정렬 : \n {np.sort(np_array,axis=0)}")

print(f"np_array 열 기준 내림차순 정렬 : \n {np.sort(np_array,axis=0)[::-1]}")
```

```python
import numpy as np
import pandas as pd
import matplotlib

# 0~ 255(미포함) 사이에서 램덤한 숫자 선택 후, 조건식 하기. 
arr1 = np.random.randint(0,255,size=(3,3))
print(f"arr1 : \n {arr1} ")

arr2 = np.random.randint(0,255,size=(3,3))
print(f"arr2 : \n {arr2} ")

#랜덤하게, 특정 값을 선택하는 형식
condArr = np.random.choice([True,False],size=(3,3))
print(f"condArr : \n {condArr}")

#조건식 걸기
np.where(condArr,arr1,arr2)
```

```python
#판다스 데이터 프레임 각 열, 행 선택 및 값 수정.

import numpy as np
import pandas as pd 
import matplotlib

data= [["상용","상용2","상용3"],
       ["메뉴1","메뉴2","메뉴3"],
       ["1000원","2000원","3000원"]]


# zip 함수로 같은 인덱스 묶기 
data2 = zip(data[0],data[1],data[2])

df = pd.DataFrame(data2, columns=["이름","메뉴","가격"])

print(f"df : \n {df} ")
# 각 열 선택 
print(f"df  이름 열 선택 : \n {df["이름"]} ")
# 각 열 값 대입
df["이름"]='lsy'
print(f"df  이름 열 값 수정 후  조회 : \n {df} ")

# 열 삭제. 
del df["이름"]
print(f"df  이름 열 값 삭제 후 조회 : \n {df} ")

# 행 선택, 수정, 삭제 
print(f"df.loc[:] 전체 행 조회 : \n {df.loc[:]} ")
print(f"df.loc[0:1] 전체 0행 조회 : \n {df.loc[0:1]} ")
# 수정 
df.loc[0] = "123"
print(f"df.loc[0] 0행 값수정. : \n {df} ")

# 특정 행에 값 추가. 행의 이름으로도 접근이 가능. 
df.loc[0,:] = ["test1","test2"]
print(f"df.loc[0,:] 0행 값 추가 후 조회. . : \n {df} ")

# iloc , 해당 인덱스의 값으로 선택, 변경이 가능함.
```
